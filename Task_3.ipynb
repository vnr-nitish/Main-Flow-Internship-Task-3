# Task-3: Data Visualization With Matplotlib

**Description:** This task involves using the matplotlib
library to visualize data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv(r"/content/householdtask3.csv")
```

```python
display(data.head(10))
```

**Scatter** **Plot**

```python
plt.scatter(data['year'],data['own'])
plt.title('Scatter plot')
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

**Line Chart**

```python
plt.plot(data['year'])
plt.plot(data['own'])
plt.title('Line chart')
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

**Bar Chart**

```python
plt.bar(data['year'],data['own'])
plt.title('Bar chart')
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

**Histogram**

```python
plt.hist(data['income'])
plt.title('Histogram')
plt.show()
```
